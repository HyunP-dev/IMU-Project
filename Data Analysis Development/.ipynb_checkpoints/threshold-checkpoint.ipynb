```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
def low_pass_filter(sequence, sampling_frequency, cut_frequency, n):
    nyq = 0.5 * sampling_frequency
    cutoff = cut_frequency / nyq
    b, a = signal.butter(n, cutoff)
    return signal.filtfilt(b, a, sequence)
```

```python
dataset = pd.read_csv("handheld_60hz_15step.csv")
dataset.head()
```

```python
l2ed = np.linalg.norm(dataset[["acc_x", "acc_y", "acc_z"]], axis=1)
plt.plot(y := low_pass_filter(l2ed, 60, 1.8, 5))

local_maximum = signal.find_peaks(y)[0]
local_minimum = signal.find_peaks(-y)[0]

plt.scatter(x=local_maximum, y=y[local_maximum])
plt.scatter(x=local_minimum, y=y[local_minimum])

plt.ylim(min(y) - 2, max(y) + 2)
plt.show()
```

```python
local_extremum = np.sort(np.append(local_maximum, local_minimum))
local_extremum
```

```python
tsdiff = np.diff(dataset.timestamp.loc[local_extremum].sort_index())
idtsdiff = [(i, d) for i, d in zip(local_extremum, tsdiff)]
idtsdiff[:5]
```

```python
sns.distplot(np.diff(dataset.timestamp.loc[local_extremum].sort_index()))
```

```python
plt.plot(y, c="black", label="Low Passed Sequence")

plt.scatter(x=local_maximum, y=y[local_maximum], c="red")
plt.scatter(x=local_minimum, y=y[local_minimum], c="blue")
plt.scatter(
    x:=list(map(lambda e: e[0], filter(lambda e: not (260<e[1]<350), idtsdiff))),
    y=[y[i] for i in x],
    c="black", label='Filtering Point'
)
plt.ylim(min(y) - 1.5, max(y) + 1.5)
plt.legend()
plt.show()
```

```python
sns.distplot(np.abs(np.diff(y[local_extremum])))
```

```python
fig, axs = plt.subplots(2, 1, gridspec_kw={'height_ratios': [4, 1]})
axs[0].scatter(x = local_extremum, y = [y[i] for i in local_extremum])
axs[0].plot(y)
axs[0].plot([9.8 for _ in y])
axs[1].plot(np.abs(np.diff([y[i] for i in local_extremum])))
axs[1].tick_params(axis='x', length=0, labelsize=0)
plt.show()
```
